```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('default')
pd.set_option('display.max_columns', None)

# üì• URLs CSV reales de Alura Store (4 tiendas)
urls = {
    'Tienda_A': 'https://raw.githubusercontent.com/Milenavila/AluraStoreLatam/main/tiendaA.csv',
    'Tienda_B': 'https://raw.githubusercontent.com/Milenavila/AluraStoreLatam/main/tiendaB.csv', 
    'Tienda_C': 'https://raw.githubusercontent.com/Milenavila/AluraStoreLatam/main/tiendaC.csv',
    'Tienda_D': 'https://raw.githubusercontent.com/Milenavila/AluraStoreLatam/main/tiendaD.csv'
}

print("üîÑ Cargando datos...")
dfs = []
for nombre, url in urls.items():
    df = pd.read_csv(url)
    df['Tienda'] = nombre
    df['Ingresos'] = df['Precio'] * df['Cantidad']
    dfs.append(df)
    print(f"‚úÖ {nombre}: {len(df)} filas")

df_all = pd.concat(dfs, ignore_index=True)
print(f"\nüìä Dataset total: {len(df_all)} ventas")
df_all.head()
```
